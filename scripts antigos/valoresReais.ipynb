### Salário

```python
path_salario_real = r'D:\UFPR\2º Semestre 2021\Estatística Econômica e Introdução à Econometria\7. Sétimo Relatório (Valores reais)\salario real.xlsx'
salario_real = pd.read_excel(path_salario_real)
```

```python
salario_real.set_index('Data', inplace=True)
```

```python
salario_real.tail()
```

```python
plt.style.use('ggplot')
ax, fig = plt.subplots(figsize=(15,8))
fig.plot(salario_real['Salario Nominal'], label='Nominal', linewidth=2)
fig.plot(salario_real['Salario Real'], label='Real', linewidth=2)
#fig.set_title('Salário Mínimo Nominal vs Salário Mínimo Real Deflacionado para Valores de Fevereiro de 2022')
fig.set_ylabel('Reais (R$)', size=12)
fig.set_xlabel('Anos', size=12)

plt.legend(
    title='Salário Mínimo:',
    title_fontsize=11,
    loc='upper left',
    ncol = 1,
    bbox_to_anchor=(0.01,0.97),
    frameon=True,
    facecolor='white',
    shadow=True,
    fontsize=10)

plt.savefig('Salário Mínimo Nominal vs Salário Mínimo Real.png', dpi=600)
```

```python
plt.style.use('ggplot')
ax, fig = plt.subplots(figsize=(16,8))
fig.plot(salario_real['IPCA Acumulado'], label='IPCA')
fig.plot(salario_real['Salario Real Acumulado'], label='Salário Mínimo Real')
fig.plot(salario_real['Salario Nominal Acumulado'], label='Salário Mínimo Nominal')
#fig.set_title('Variação do IPCA Acumulada no Período (1994-2022)')
fig.set_ylabel('Variação (%)', size=12)
fig.set_xlabel('Meses', size=12)

plt.legend(
    title='Variação Acumulada:',
    title_fontsize=11,
    loc='upper left',
    ncol = 1,
    bbox_to_anchor=(0.01,0.97),
    frameon=True,
    facecolor='white',
    shadow=True,
    fontsize=10)

#plt.savefig('Salário Mínimo Real vs IPCA Acumulado no Período.png', dpi=600)
```

```python
diff_real_nominal = salario_real.filter(['Salario Nominal', 'Salario Real'])
```

```python
diff_real_nominal['Diferença'] = diff_real_nominal['Salario Real'] - diff_real_nominal['Salario Nominal'] 
```

```python
plt.style.use('ggplot')
ax, fig = plt.subplots(figsize=(16,8))
fig.plot(diff_real_nominal['Diferença'], label='Diferença entre salário real e nominal')

fig.set_ylabel('Diferença (R$)', size=12)
fig.set_xlabel('Meses', size=12)

plt.legend(
    title='Variação:',
    title_fontsize=11,
    loc='upper left',
    ncol = 1,
    bbox_to_anchor=(0.01,0.97),
    frameon=True,
    facecolor='white',
    shadow=True,
    fontsize=10)

plt.savefig('Diferença entre Salário Nominal e Salário Real.png', dpi=600)
```
